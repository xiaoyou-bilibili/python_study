```python
# 一个最简单的例子
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait

# 使用谷歌浏览器
browser = webdriver.Chrome()
try:
    # 访问百度
    browser.get("https://wwww.baidu.com")
    # 获取输入框
    input = browser.find_element_by_id('kw')
    # 发送内容
    input.send_keys('Python')
    # 按回车
    input.send_keys(Keys.ENTER)
    # 等待响应
    wait = WebDriverWait(browser, 10)
    # 打印结果
    wait.until(EC.presence_of_element_located((By.ID, 'content_left')))
    # 打印链接信息
    print(browser.current_url)
    print(browser.get_cookie())
    print(browser.page_source)
finally:
    pass
    # browser.close()
```

```python
# 我们可以使用get传入页面
from selenium import  webdriver

browser = webdriver.Chrome()
# 手动打开淘宝页面
browser.get('https://www.taobao.com')
print(browser.page_source)
# browser.close()
```

```python
# 可以使用find来查找节点
# 根据id来查找节点
input_first = browser.find_element_by_id('q')
# 根据css来查找节点
input_second = browser.find_element_by_css_selector('#q')
# 根据xpath来查找节点
input_third = browser.find_element_by_xpath('//*[@id="q"]')
# 最后打印一下结果，可以看到这三个结果是一样的
print(input_first)
print(input_second)
print(input_third)
```

```python
# find_element_by_id的等价方法就是下面这个
input_first = browser.find_element(By.ID, 'q')
print(input_first)
```

```python
# 如果有多个节点那么就需要使用find_elements
lis = browser.find_elements_by_css_selector('.service-bd li')
print(lis)
```

```python
# 找到了节点后我们就可以开始进行节点交互了
# 比如这里我们找到输入框然后输入对应的字符
import time
input = browser.find_element_by_id('q')
# 输入iphone
input.send_keys('iphone')
time.sleep(1)
# 清空
input.clear()
input.send_keys('ipad')
# 找到搜索按钮
button = browser.find_element_by_class_name('btn-search')
# 点击
button.click()
```

```python
# 上面只是简单的展示了点击和输入，下面来执行一些复杂的操作
from selenium.webdriver import ActionChains

browser = webdriver.Chrome()
# 打开演示界面
browser.get('https://www.runoob.com/try/try.php?filename=jqueryui-api-droppable')
# 切换到对应的frame
browser.switch_to.frame('iframeResult')
# 找到待拖拽的对象
source = browser.find_element_by_css_selector('#draggable')
# 找到对应的目标
target = browser.find_element_by_css_selector('#droppable')
# 打开一个action链
actions = ActionChains(browser)
# 拖动对象
actions.drag_and_drop(source,target)
# 执行
actions.perform()
```

```python
# 我们还可以执行JavaScript脚本
browser = webdriver.Chrome()
browser.get('https://www.zhihu.com/explore')
# 执行脚本，滚动到底部
browser.execute_script('window.scrollTo(0,document.body.scrollHeight);')
# 显示alert弹窗
browser.execute_script('alert("to bottom")')
```

```python
# 下面我们来提取一下节点的信息
browser = webdriver.Chrome()
browser.get('https://www.zhihu.com/explore')
# 我们获取一下logo向欧盟西欧
logo = browser.find_element_by_class_name('ExploreHomePage-specialsLoginImg')
print(logo)
# 获取对应的class值
print(logo.get_attribute('class'))
```

```python
# 我们还可以获取其他属性
# 打印id信息
print(input.id)
# 打印对应的位置
print(input.location)
# tag名称
print(input.tag_name)
# 标签的大小
print(input.size)
```

```python
# 如果网站需要等待一段时间后才能加载出来，这个时候我们就需要触发延时等待了
browser = webdriver.Chrome()
# 隐式等待10s
browser.implicitly_wait(10)
browser.get('https://www.zhihu.com/explore')
input = browser.find_element_by_class_name('Input')
print(input)
```

```python
# 隐式等待效果并不是很好，这个只是规定了一个固定的时间，而页面加载时间会受到网络条件的影响
# 下面我们进行显式的等待

browser = webdriver.Chrome()
browser.get('https://www.taobao.com/')
# 这里设置一个最长的等待时间，我们只等待10s
wait = WebDriverWait(browser, 10)
# 等待搜索框的出现
input = wait.until(EC.presence_of_element_located((By.ID,'q')))
button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.btn-search')))
print(input)
print(button)
```

```python
# 我们还可以进行前进和后退操作
browser = webdriver.Chrome()
# 我们先依次访问三个网站
browser.get('https://www.baidu.com')
browser.get('https://www.taobao.com')
browser.get('https://www.python.org')
# 后退
browser.back()
# 等待1s
time.sleep(1)
# 前进
browser.forward()
```

```python
# 下面演示一下使用cookie
browser = webdriver.Chrome()
browser.get('https://www.zhihu.com/explore')

# 打印cookie信息
print(browser.get_cookies(),sep='\n\n')
# 添加cookie信息
browser.add_cookie({'name':'name','domain':'www.zhihu.com','value':'germey'})
# 重新获取一下cookie信息
print(browser.get_cookies(),sep='\n\n')
# 删除所有cookie
browser.delete_all_cookies()
print('新的cookie')
print(browser.get_cookies(),sep='\n\n')
```

```python
# 选项卡管理
browser = webdriver.Chrome()
browser.get('https://www.baidu.com')
# 这里打开一个新的窗口
browser.execute_script('window.open()')
# 打印所有的窗口
print(browser.window_handles)
# 切换到新的选项卡
browser.switch_to.window(browser.window_handles[1])
browser.get('https://www.taobao.com')
time.sleep(1)
# 我们再切换到第一个
browser.switch_to.window(browser.window_handles[0])
browser.get('https://python.org')
```

```python
# 引入异常信息
from selenium.common.exceptions import TimeoutException, NoSuchElementException
# 最后就是异常处理
browser = webdriver.Chrome()
try:
    browser.get('https://www.baidu.com')
except TimeoutException:
    print('time out')

try:
    browser.find_element_by_id('hello')
#  没有找到这个错误的异常
except NoSuchElementException:
    print('no element')
finally:
    browser.close()
```
