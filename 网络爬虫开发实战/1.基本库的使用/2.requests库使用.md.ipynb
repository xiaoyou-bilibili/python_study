## 基本用法

```python
import requests

r = requests.get('https://www.baidu.com')
print(type(r))
print(r.status_code)
print(type(r.text))
print(r.text)
print(r.cookies)
```

我们可以快捷的使用其他请求

![](.1_images/e4e3ee26.png)

```python
# 下面我们使用get请求然后携带参数
r = requests.get('http://httpbin.org/get?name=germey&age=22')
print(r.text)
```

```python
# 上面这个使用起来其实没那么方便，我们可以这样
data = {
    'name': 'germey',
    'age':22
}
r = requests.get('http://httpbin.org/get', params=data)
print(r.text)
```

```python
# 下面我们直接转换成字典类型数据
r = requests.get('http://httpbin.org/get')
print(type(r.text))
print(r.json())
print(type(r.json()))
```

```python
# 下面我们简单爬一下知乎
import requests
import re

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36 Edg/102.0.1245.33'
}

r = requests.get('https://www.zhihu.com/explore', headers=headers)
# 这里使用正则来替换
pattern = re.compile('class="css-1nd7dqm".*?>(.*?)</a>', re.S)
title = re.findall(pattern, r.text)
print(title)
```

```python
# 下面我们爬取二进制的数据
r = requests.get('https://github.com/favicon.ico')
# 二进制使用文本表示会乱码
print(r.text)
# 使用content才能正常表示二进制的数据
print(r.content)
```

```python
# 下面我们保存一下这个图片
with open('favicon.ico', 'wb') as f:
    f.write(r.content)
```

```python
# 下面判断一下post请求
data = {'name':'germey', 'age': 22}
r = requests.post('http://httpbin.org/post', data=data)
print(r.text)
```

```python
# 获取一下响应结果
r = requests.get('http://www.jianshu.com')
print(type(r.status_code), r.status_code)
print(type(r.headers), r.headers)
print(type(r.cookies), r.cookies)
```

```python
# 下面这里有一些常用的状态码
print(requests.codes.ok)
```

## 高级用法

```python
# 下面我们来模拟上传文件
files = {'file': open('favicon.ico', 'rb')}
r = requests.post('http://httpbin.org/post', files=files)
print(r.text)
```

```python
# 下面获取一下cookie数据
r = requests.get('https://www.baidu.com')
print(r.cookies)
```

```python
# 遍历cookie数据
for key,value in r.cookies.items():
    print('%s=%s' % (key,value))
```

```python
# 我们每次请求都携带cookie比较麻烦，我们可以使用会话功能
s = requests.Session()
s.get('http://httpbin.org/cookies/set/number/123456789')
r = s.get('http://httpbin.org/cookies')
print(r.text)
```

```python
# 我们请求的时候可以不验证SSL证书
r = requests.get('', verify=False)
```

```python
# 下面使用代理
proxies = {
    'http': 'http://192.168.1.1:7890'
}
r = requests.get('https://google.com', proxies=proxies)
print(r.text)
```

```python
# 我们也可以使用socks代理，需要先安装库
# pip install requests[socks]
proxies = {
    'http': 'socks5://user:password@host:port'
}
r = requests.get('https://google.com', proxies=proxies)
```

```python
# 设置超时
r = requests.get('https://www.baidu.com', timeout=1)
print(r.status_code)
```

```python
# 请求分为连接和读取，如果想分别指定那么就可以传入一个元组
r = requests.get('https://www.baidu.com', timeout=(5,11))
# 如果想永久等待，那么可以这样
r = requests.get('https://www.taobao.com', timeout=None)
```

```python
# 下面演示一下身份认证
r = requests.get('http://192.168.123.1', auth=('username', 'passwd'))
```

```python
# 当然还支持其他认证方式，不过需要下载对应的包
# 如果是oauth认证，那么就需要先安装oauth包
# pip3 install requests_oauthlib
```

```python
# 下面我们可以查看一下预处理对象
from  requests import Request,Session

url = 'http://httpbin.org/post'
data = {
    'name': 'germey'
}
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36 Edg/102.0.1245.33'
}
s = Session()
req = Request('POST', url, data=data, headers=headers)
prepped = s.prepare_request(req)
r = s.send(prepped)
print(r.text)
```
